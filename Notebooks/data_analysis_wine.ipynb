```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_df = pd.read_csv("winequality-red.csv", delimiter=';')
```

```python
data_df
```

```python
# Tipos de qualidades do vinho vermelho
data_df["quality"].unique()
```

```python
# Lista de 4 atributos selecionados
lista_atributos = ["fixed acidity", "residual sugar", "pH", "alcohol"]
```

```python
for i in lista_atributos:
    print(i+": ")
    print("Media = "+str(data_df[i].mean()))
    print("Moda  = "+str(data_df[i].mode()))
    print("Mediana  = "+str(data_df[i].median()))
    print("Percentis 20, 50 e 70 = " + str(np.percentile(data_df[i], [20, 50, 70])))
    print("Quartis Q1, Q2 e Q3 = " + str(np.quantile(data_df[i], [0.25, 0.5, 0.75])))
    print("VariÃ¢ncia  = "+str(data_df[i].var()))
    print("Desvio PadrÃ£o  = "+str(data_df[i].std()))
    print("")
```

```python
df_new = pd.DataFrame(data_df.reindex(columns = (['fixed acidity', 'residual sugar', 'pH', 'alcohol'])))

for i in lista_atributos:
    df_new[i] = df_new[i]/df_new[i].max()
    
```

```python
plt.figure(figsize=(10,7))
plt.style.use('seaborn')
df_new.boxplot(column=["fixed acidity", "residual sugar", "pH", "alcohol"])
plt.savefig("boxplot.png", bbox_inches='tight')
plt.show()
```

```python
df_new = pd.DataFrame(data_df.reindex(columns = (['fixed acidity', 'residual sugar', 'pH', 'alcohol'])))
```

```python

plt.figure(figsize=(15, 15))
plt.subplot(4,2,1), plt.hist(df_new["fixed acidity"])
plt.title('Fixed acidity', size = 15), plt.xlabel("Valor", size = 10,), plt.ylabel("Quantida", size = 10)
plt.subplot(4,2,2),plt.hist(df_new["residual sugar"])
plt.title('Residual sugar', size = 15), plt.xlabel("Valor", size = 10,), plt.ylabel("Quantida", size = 10)
plt.subplot(4,2,3),plt.hist(df_new["pH"])
plt.title('pH', size = 15), plt.xlabel("Valor", size = 10,), plt.ylabel("Quantida", size = 10)
plt.subplot(4,2,4),plt.hist(df_new["alcohol"])
plt.title('Alcohol', size = 15), plt.xlabel("Valor", size = 10,), plt.ylabel("Quantida", size = 10)
plt.tight_layout()
plt.savefig("histograma.png", bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(12, 12))
plt.subplot(2,2,1), plt.scatter(df_new["fixed acidity"], df_new["alcohol"])
plt.title('Fixed acidity x Alcohol', size = 15), plt.xlabel("fixed acidity", size = 10,), plt.ylabel("alcohol", size = 10)
plt.subplot(2,2,2),plt.scatter(df_new["residual sugar"], df_new["pH"])
plt.title('Residual sugar x pH', size = 15), plt.xlabel("residual sugar", size = 10,), plt.ylabel("pH", size = 10)
plt.tight_layout()
plt.savefig("acatt.png", bbox_inches='tight')
plt.show()
```
